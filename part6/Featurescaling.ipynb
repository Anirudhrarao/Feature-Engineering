#### FEATURE SCALING

```python
# Loading the dependencies
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

#### 1) STANDARDIZATION

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
```

```python
standardize_data = []
for i in list(df['total_bill']):
    z = (i-mean)/std
    standardize_data.append(z)
standardize_data[:5]
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(standardize_data)
```

As you can see distribution of data is same but we changed the scaling.

#### Standardization is also available in scikit learn

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df['total_bill'] = scaler.fit_transform(df[['total_bill']])
```

```python
df['total_bill'].head()
```

#### 2) NORMALIZATION MIN-MAX-SCALING

```python
# loading dataset
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
min = np.min(df['fare'])
max = np.max(df['fare'])
```

```python
normalized_data = []
for i in list(df['fare']):
    norm = (i - min)/(max - min)
    normalized_data.append(norm)
normalized_data[:5]
```

#### Normalization is also available in scikit learn

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
df['fare'] = minmax.fit_transform(df[['fare']])
```

```python
df['fare'].head()
```

It will scale down between 0 to 1.
