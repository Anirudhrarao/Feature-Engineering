### SMOTE TECHNIQUE IMPLEMENTATION:

```python
# import lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_classification
```

```python
# create dataset
X,y = make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],n_redundant=0,random_state=12)
```

```python
# create dataframe
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df = pd.concat([df1,df2],axis=1)
```

```python
# read the data
df.head()
```

```python
# value count fot target
df['target'].value_counts()
```

```python
plt.scatter(x=df['f1'],y=df['f2'],c=df['target'],cmap='coolwarm')
plt.show()
```

### SMOTE - (oversampling)

```python
# import lib
from imblearn.over_sampling import SMOTE
```

```python
# create instance for SMOTE class
oversample = SMOTE()
```

```python
# oversample
X,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
# check shape for X and y
X.shape,y.shape
```

```python
# create dataframe
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df = pd.concat([df1,df2],axis=1)
```

```python
# read data
df.head()
```

```python
# scatter plot
plt.scatter(df['f1'],df['f2'],c=df['target'])
plt.show()
```
