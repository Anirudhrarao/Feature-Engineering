### PERCENTILE AND QUARTILES USING NUMPY

```python
import numpy as np
age = [20, 2, 7, 1, 34]
print('25th percentile of age:',np.percentile(age,25))
print('50th percentile of age:',np.percentile(age,50))
print('75th percentile of age:',np.percentile(age,75))
```

```python
import numpy as np
age = [20, 2, 7, 1, 34]
Q1 = np.quantile(age,0.25)
Q2 = np.quantile(age,0.50)
Q3 = np.quantile(age,0.75)
print(Q1,Q2,Q3)
```

### 5 NUMBER SUMMARY AND BOXPLOT FOR OUTLIERS

```python
import numpy as np
marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000]
min,Q1,median,Q3,max = np.quantile(marks,[0,0.25,0.5,0.75,1])
```

```python
min,Q1,median,Q3,max
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_fence = Q1 - 1.5*(IQR)
lower_fence
```

```python
higher_fence = Q3 + 1.5*(IQR)
higher_fence
```

#### The value below 1.5 and value above 141.5 consider as outlier in data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# set size
plt.figure(figsize=(4,2))
# Set the color palette
colors = sns.color_palette('Set2')
# Create the boxplot
ax = sns.boxplot(marks, palette=colors, width=0.3,orient='h')
# title
plt.title('Boxplot of marks')
plt.show()
```

```python
# let's print outliers
for i in marks:
    if (i>higher_fence) or (i<lower_fence):
        print(f'Outlier = {i}')
```

```python
# removing outliers from marks
for x in marks:
    if (i>higher_fence) or (i<lower_fence):
        marks.remove(x)
marks
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# set size
plt.figure(figsize=(4,2))
# Set the color palette
colors = sns.color_palette('Set2')
# Create the boxplot
ax = sns.boxplot(marks, palette=colors, width=0.3,orient='h')
# title
plt.title('Boxplot of marks')
plt.show()
```
